# Exercise 3.1: Plotting experimental data

In the [online resources](http://www-personal.umich.edu/~mejn/cpresources.zip) you will find a file called sunspots.txt, which contains the observed number of sunspots on the Sun for each month since January 1749. The file contains two columns of numbers, the first being the month and the second being the sunspot number.

a)

Write a program that reads in the data and makes a graph of sunspots as a function of time

```python
from numpy import loadtxt
import matplotlib.pyplot as plt
import math
```

```python
data = loadtxt("sunspots.txt", float)
time = [i[0] for i in data]
sunspots = [i[1] for i in data]

plt.figure(figsize=(10, 6))
plt.plot(time, sunspots, 'b.')
plt.show()
```

b)
Modify your program to display only the first 1000 data points on the graph.

```python
plt.figure(figsize=(10, 6))
plt.plot(time[:1000], sunspots[:1000])
plt.show()
```

c)
Modify your program further to calculate and plot the running average of the data, defined by

$$ Y_k = \frac{1}{2r}\sum^r_{m=-r}y_{k+m},$$

where $r=5$ in this case (and the $y_k$ are the sunspot numbers). Have the program plot both the original data again over the range covered by the first 1000 data points.

```python
def running_average(y, r=5):
    '''
    Calculates the running average for the number of sunspots with a window size of 2*r
    parameters:
    y   - list of floats data to be averaged
    r   - int the window size parameter
    avg - the averaged result from the window function
    
    '''
    avg = []
    weight = (2*r+1)
    for i in range(len(y)-2*r):
        weighted = sum(y[i:i+2*r])/weight
        avg.append(weighted)
    return avg

```

```python
weighted = running_average(sunspots)
plt.figure(figsize=(10, 6))
plt.plot(time[:1000], sunspots[:1000],'r.', alpha=0.5)
plt.plot(time[5:1000-5], weighted[:990],'k')
plt.show()
```
