Create a user-defined function f(x) that returns the value $1+\frac{1}{2}tanh2x$, then use a central difference to calculate the derivative of the functionin the range, $-2 \leq x \leq 2$. Calculate an analytic formula for the derivative and make a graph with your numerical result and the analytic answer on the same plot. It may help to plot the exact answer as line and the numerical one as dots. (Hint: In Python the ntanh function is found in the math package, and it's called simpy tanh.)

```python
import numpy as np
import matplotlib.pyplot as plt
from math import tanh
```

```python
def f(x):
    return 1 + np.tanh(2*x)/2
```

```python
def der_analytical(x):
    return  1/np.cosh(2*x)**2#1 - np.tanh(2*x)**2 
```

```python
def der_numerical(f, x, h):
    return (f(x+h) - f(x-h))/(2*h)
```

```python
h = 1e-3


x = np.linspace(-2, 2, 401)
analytical = der_analytical(x)
numerical = der_numerical(f,x,h)



plt.figure(figsize=(10,6))
plt.title("The derivative of f(x) analytically and numerically")
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, analytical, label="Analytical")
plt.plot(x, numerical, 'r.', alpha=0.3, label="Central difference")
#plt.plot(x, f(x))
plt.legend()
plt.plot()
```
