# Exercise 5.13: Quantum uncertainty in the harmonic oscillator

In units where all the constant are 1, the wavefunction of the $n$th energy level of the one-dimensional quantum harmonic oscillator -- i.e., a spinless point particle in a quadratic potential well -- is given by

$$\psi_n(x) = \frac{1}{\sqrt{2^nn!\sqrt{\pi}}}e^{-x^2/2}H_n(x)$$

for $n=0\ldots\infty,$ where $H_n(x)$ is the $n$th Hermite polynomial. Hermite polynomials satisfy a relation somewhat similar to that for the Fibonacci numbers, although more complex:

$$H_{n+1}(x) =2xH_n(x)-2nH_{n-1}(x).$$

The first two Hermite polynomials are $H_0(x)=1$ and $H_1(x)=2x$.

A) Write a user-defined function H(n,x) that calculates $H_n(x)$ for given $x$ and any integer $n\geq0$. Use your function to make a plot that shows the harmonic oscillator wavefunctions for $n=0,1,2,\text{and }3$, all on the same graph, in the range $x=-4$ to $x=4$. Hint: There is a a function factorial in the math package that calculates the factorial of an integer.

```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

```python
def H(n,x):
    if n==0:
        return 1
    if n==1:
        return 2*x
    return 2*x*H(n-1,x) - 2*(n-1)*H(n-2,x)
```

```python
def psi(n,x):
    hermite_part = H(n,x)
    other_part = 1/np.sqrt(2**n * factorial(n) * np.sqrt(np.pi))*np.exp(-x**2/2)
    return other_part*hermite_part
```

```python
wavefunctions = [[] for _ in range(4)]

plt.figure(figsize=(10,6))
plt.title("The wave function for the quantum harmonic oscillator")
plt.xlabel("x")

for n in range(4):
    for x in np.linspace(-4, 4, 800):
        p = psi(n,x)
        wavefunctions[n].append(p)
    plt.plot(np.linspace(-4,4,800), wavefunctions[n], label=f"n={n}")
plt.legend()
plt.show()
```
